```python
# 12. Use a for loop to iterate through the string "Hello" and print each character separated by a hyphen (e.g., H-e-l-l-o).

str4 = "Hello"

for index, i in enumerate(str4):
    if index == len(str4)-1:
        print(i)
    else:
        print(i,end="-")
```
