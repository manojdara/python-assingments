```python
class t-shirt:
    def __init__ (self,material,size,pattern,color,tag):
        self.material = material
        self.size = size
        self.pattern = pattern
        self.color = color
        self.tag = tag

    def stitching(self):
        print(self.material,"is the material used in making the shirt ")
        
    def joining(self):
        print(self.pattern,"is the pattern waived into the shirt")

    def pressing(self):
        print("Press the shirt with Brand logo")

    def cleaning(self):
        print("clean the shirt")

    def ironing(self):
        print("Iron the shirt")
```
