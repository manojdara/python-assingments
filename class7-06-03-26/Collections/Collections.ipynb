### Below are the List of Python Collections
1. List
2. Tuple
3. Set
4. Dictionary

```python
#List - CURD (Create, Update, Read, and Delete)

# Create the List 
l1=list()
l2 = [1,2,3]

print(type(l1))
print(type(l2))


#Update a List

l2.append("Mango")
l2.extend([4,5])
l2.insert(1,0)

#Read the list
print(l2)
print(l2[4])
```

```python
#Delete an element 

l2.remove(1)
print(l2)
```

```python
#Tuple

#Create

t1 = tuple()
t2 = (1,2,3,4)

#Update in tuple is not possible as it is immutable data type. Need to convert to list and make the changes

list1 = list(t2)
list1[0] = 0
t2 = tuple(list1)

print(t2)
print(t2[0])

#Read 

print(t2)
print(type(t2))

#Delete
del t2 #If try to print will get a name error
```

```python
#Set

#Create

set1 =set()
set2 = {"Dosa","Idly","Vada"}

#Read
print(set2)
print("Idly" in set2)

#Update
set2.add("Poori")
print(set2)

#Delete
set2.remove("Vada")
print(set2)
set2.discard("Vada") #This method does not throw out any error even if element is not present in the set.
print(set2)
```

```python
#Dictionary 

#Create
leo=dict()
marks = { "Maths":85, "English": 75, "Science": 80, "EVS": 90, "Telugu": 88, "History": 81 }

#Read
print(type(leo))
print(marks)
print(marks["Maths"])

#Update

marks["Commerce"] = 82
print(marks)

#Delete
del marks["Telugu"]
print(marks)
```
