```python
# 1. Write a program to print the multiplication table of a given number.For Loop

n = int(input("Enter a number for multiplication table"))

for i in range(11):
    prod = n*i
    print(f"{n}x{i}={prod}")
```

```python
# 2. Write a program to print the Fibonacci series up to a given number.

n = int(input("Enter the number of terms: "))

a = 0
b = 1

print("Fibonacci Series:")

for i in range(n):
    print(a, end=" ")
    c = a + b
    a = b
    b = c
```

```python
# 3. Write a program to find the factorial of a given number.

n = int(input("Enter the number of terms: "))
fact = 1
for i in range(1, n+1):
    fact = fact*i

print(f"Factorial of {n} is {fact}")
```

```python
# 4. Write a program to print the first n prime numbers.

for num in range(2, 101):
    is_prime = True

    for i in range(2, num):
        if num % i == 0:
            is_prime = False
            break

    if is_prime:
        print(num, end=" ")
```

```python
# 5. Write a program to find the sum of all natural numbers from 1 to n.
n = int(input("Enter a number: "))

total = 0

for i in range(1, n + 1):
    total = total + i

print(f"Sum of natural numbers from 1 to {n} is {total}")
```

```python
# 6. Write a program to print the ASCII values of all the characters in a string.

text = input("Enter a string: ")

for ch in text:
    print(f"{ch} : {ord(ch)}")
```

```python
# 9. Write a program to print the reverse of a string.

s = input("Enter a string: ")

reverse_str = ""

for ch in s:
    reverse_str = ch + reverse_str

print("Reversed String:", reverse_str)
```

```python
# 10. Write a program to print the reverse of a list.

l1 = [10, 20, 30, 40, 50]

reverse_list = []

for i in range(len(l1)-1,-1,-1):
    reverse_list.append(l1[i])

print("Original List:", l1)
print("Reversed List:", reverse_list)
```

```python
# 11. Write a program to find all the prime numbers between a given range using for loop.
start = int(input("Enter the start of the range: "))
end = int(input("Enter the end of the range: "))

print("Prime Numbers:")

for num in range(start, end + 1):

    if num < 2:
        continue

    is_prime = True

    for i in range(2, num):
        if num % i == 0:
            is_prime = False
            break

    if is_prime:
        print(num, end=" ")
```
