```python
#1. Print numbers from 1 to 10 using a while loop.

n=1

while(n<=10):
    print(n)
    n=n+1
```

```python
# 2. Print numbers from 10 to 1 using a while loop.

n=10
while(n>0):
    print(n)
    n=n-1
```

```python
# 3. Print all even numbers from 2 to 20 using a while loop.

n = 2

while(n<21):
    if(n%2 == 0):
        print(n)
        n=n+1
    else:
        n=n+1
        continue
```

```python
# 4. Print all odd numbers from 1 to 19 using a while loop.

n = 1

while(n<20):
    if(n%2 != 0):
        print(n)
        n=n+1
    else:
        n=n+1
        continue
```

```python
#5. Find the sum of numbers from 1 to n using a while loop.

n = int(input("Enter a length of a number range: "))
i=1
total=0
while(i<=n):
        total=total+i
        i=i+1


print("The total sum of the numbers is: ", total)
        
```

```python
#6. Print the multiplication table of a given number using a while loop.

n=int(input("Enter the number for it's multiplication Table: "))
print("Multiplication Table for: ",n)
i=0

while(i<=10):
    mul=n*i
    print("{}X{}={}".format(n,i,mul))
    i=i+1
```

```python
#7. Count the number of digits in a given number using a while loop.

n = int(input("Enter any number: "))
count = 0
while(n!=0):
    count = count + 1
    n=n//10
    
print(count)
```

# Beginner Nested For Loop Questions
1. Print Hello 3 times in 3 rows using nested loops.
2. Print the pattern: * * * on two rows.
3. Print the pattern: 1 1 1 on two rows.
4. Print numbers from 1 to 3 in each row for 3 rows.
5. Print the pattern: A B C on two rows.
6. Print a 4 × 2 rectangle of stars.
7. Print all pairs of numbers from 1 to 2.
8. Print a 3 × 4 rectangle of X characters.
9. Print the row and column numbers for a 2 × 3 grid.
10. Print the pattern: 1 2 on four rows.

```python
# 1. Print Hello 3 times in 3 rows using nested loops.

str = "Hello"


for i in range(0,3):
    for j in range(0,3):
        print(str,end=" ")
    print()
```

```python
# 2. Print the pattern: * * * on two rows.

for i in range(0,2):
    for j in range(0,3):
        print("*",end=" ")
    print()
```

```python
# 3. Print the pattern: 1 1 1 on two rows.
for i in range(0,2):
    for j in range(0,3):
        print("1",end=" ")
    print()
```

```python
# 4. Print numbers from 1 to 3 in each row for 3 rows.
n=1
for i in range(0,3):
    for j in range(0,3):
        print(n,end=" ")
        n=n+1
    print()
    n=1
```

```python
# 5. Print the pattern: A B C on two rows.
pat = ["A","B","C"]
for i in range(0,2):
    for j in range(0,3):
        print(pat[j],end=" ")
    print()
```

```python
# 6. Print a 4 × 2 rectangle of stars.

for i in range(0,4):
    for j in range(0,2):
        print("*",end=" ")
    print() 
```

```python
# 7. Print all pairs of numbers from 1 to 2.

for i in range(1,3):
    for j in range(1,3):
        print(f"({i},{j})",end= " ")
    print()
```

```python
# 8. Print a 3 × 4 rectangle of X characters.

for i in range(0,3):
    for j in range(0,4):
        print("X",end = " ")
    print()
```

```python
# 9. Print the row and column numbers for a 2 × 3 grid.

for i in range(0,2):
    for j in range(0,3):
        print(f"({i},{j})",end=" ")
    print()
```

```python
# 10. Print the pattern: 1 2 on four rows.

for i in range(0,4):
    for j in range(1,3):
        print(f"{j}",end=" ")
    print()
```
