```python
# class and Objencts


class ps4_game:
    def __init__ (self,game_name,cd,noofplayers,length,noofcontrollers):
        self.game_name = game_name
        self.cd=cd
        self.noofplayers = noofplayers
        self.length = length
        self.noofcontrollers = noofcontrollers

    def gname(self):
        self.game_name = "Devil May cry 4"
        print(self.game_name)

    def tutorial(self):
        print(self.game_name," Game Tutorial")

    def action_sequence(self):
        print("Action Sequence")

    def save_point(self):
        print("Game Save Point")

    def reload(self):
        print("Load Game from last save point")


p = ps4_game("New Disc","DMc3","1","600","1")
p.gname()
p.tutorial()
p.action_sequence()
p.save_point()
p.reload()
print("Let's start")
```

```python
# Inheritance: Inheritance allows us to define a class that inherits all the properties and methods of another class.

# There are 5 different types of Inheritances which are as follows:
# 1. Single Inheritance: A child class inherits from one parent class.
# 2. Multiple Inheritance: A child class inherits from more than one parent class. 
# 3. Multilevel Inheritance: A class inherits from a class that already inherits from another class. 
# 4. Hierarchical Inheritance: Multiple child classes inherit from the same parent class.
# 5. Hybrid Inheritance: A combination of two or more inheritance types.

# Single Inheritance Example

class py_class:
    def greet(self):
        print("Good Morning All")
        
class student(py_class):
    def question(self):
        print("What are concepts that we are going to be learning today")

s = student()
s.greet()
s.question()
print("Thank you for the response")
```

```python
# Multiple Inheritance: A child class inherits from more than one parent class. 
class py_class:
    def greet(self):
        print("Good Morning All")
        
class missed_class:
    def video_recordings(self):
        print("Video recordings for all the classes are available")
        
class student(py_class,missed_class):
    def question(self):
        print("What are concepts that we are going to be learning today")

s = student()
s.greet()
print("If you missed the class go ahead and view the video recordings")
s.video_recordings()
```

```python
# Multilevel Inheritance: A class inherits from a class that already inherits from another class.

class py_class:
    def greet(self):
        print("Good Morning All")
        
class student(py_class):
    def question(self):
        print("What are concepts that we are going to be learning today")

class presentation(student):
    def display(self):
        print("Please wait for the presentation")

p = presentation()
p.greet()
p.display()
print("Thank you for your patience")
```

```python
# Hierarchical Inheritance: Multiple child classes inherit from the same parent class.

class py_class_recordings:
    def recorded(self):
        print("class videos are raedy")
        
class batch1(py_class):
    def view(self):
        print("Need to go through the videos")
        
class batch2(py_class):
    def view(self):
        print("Need to go through the videos for batch 2")

b = batch1()
b.view()
b2 = batch2()
b2.view()
        
```

```python
# Hybrid Inheritance: A combination of two or more inheritance types.
class PyClass:
    def greet(self):
        print("Good Morning All")

class Student(PyClass):
    def question(self):
        print("What are concepts that we are going to be learning today")

class Assignment(PyClass):
    def submission(self):
        print("Assignment Submitted")

class Presentation(Student, Assignment):
    def display(self):
        print("Time for the presentation to start")

p = Presentation()

p.greet()
p.question()
p.submission()
p.display()
```

```python
# Polymorpshim: It is OOP concept where the same method,function can behave diffenertly depending on the 
# object or data type that is being referenceed.

class Teacher:
    def work(self):
        print("Teaching students")

class Doctor:
    def work(self):
        print("Treating patients")

t = Teacher()
d = Doctor()

t.work()
d.work()
```

```python
# Encapsulation: Binding variables(attributes) and methods(functions) together in class and restricting direct access for some data.
# Data + Methods + Protection = Encapsulation

class Student:
    def __init__(self, name, marks):
        self.name = name
        self.__marks = marks  

    def set_marks(self, marks):
        if 0 <= marks <= 100:
            self.__marks = marks
        else:
            print("Marks should be between 0 and 100")

    def get_marks(self):
        return self.__marks

s1 = Student("Manoj", 65)

print("Marks:", s1.get_marks())

print("After Revaluation")

s1.set_marks(75)
print("Updated Marks:", s1.get_marks())
```

```python
# Abstration: It is the process of hiding the implementation details and exposing the required functionality to the user.

class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def get_percentage(self):
        total = sum(self.marks)
        percentage = (total / 600) * 100
        return percentage
    
    def display_student_details(self):
        print("Student Name",self.name)
        print("Student Age", 30)

s1 = Student("Kevin", [80, 85, 90, 75, 95, 88])
s1.display_student_details()
print("Percentage:", s1.get_percentage())



```

## What is a decorator?
A decorator is a function that takes another function as input, adds additional functionality to it, and returns the modified function without changing the original function's code.

```python
is_logged_in = True

def login_required(func):

    def wrapper():
        if is_logged_in:
            func()
        else:
            print("Please Login First")

    return wrapper


@login_required
def view_dashboard():
    print("Welcome to Student Dashboard")


view_dashboard()
is_logged_in = False
view_dashboard()
```
